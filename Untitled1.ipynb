```python
import re
import io
import PyPDF2
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
from google.colab import files
from transformers import T5ForConditionalGeneration, T5Tokenizer

print("✅ Imports ready!")
```

```python
# Using FLAN-T5-large — much better than base for rewriting tasks
print("⏳ Loading FLAN-T5-large... (takes 1-2 min first time)")

MODEL_NAME = "google/flan-t5-large"
tokenizer = T5Tokenizer.from_pretrained(MODEL_NAME)
model = T5ForConditionalGeneration.from_pretrained(MODEL_NAME)

print("✅ Model ready!")
```

```python
# ── Rewrite one section ───────────────────────────────────────────────────────
def rewrite_section(label, content, job_role, jd_keywords):
    """
    Rewrites a single resume section using FLAN-T5-large.
    label     : e.g. "OBJECTIVE", "SKILLS", "PROJECTS"
    content   : the actual text under that section
    job_role  : e.g. "Software Tester"
    jd_keywords: top keywords pulled from JD
    """
    if not content.strip():
        return content

    kw = ", ".join(jd_keywords[:12])

    prompt = (
        f"Rewrite this resume {label} section for a {job_role} job application. "
        f"Keep all original facts. Use action verbs. "
        f"Include these keywords naturally: {kw}. "
        f"Be concise and professional. Preserve the original structure.\n"
        f"Original: {content[:500]}\n"
        f"Rewritten:"
    )

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        max_length=600,
        truncation=True
    )

    outputs = model.generate(
        inputs["input_ids"],
        max_new_tokens=300,
        num_beams=5,
        early_stopping=True,
        no_repeat_ngram_size=3,
        length_penalty=1.5
    )

    result = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    return result


# ── Extract keywords from JD ──────────────────────────────────────────────────
def get_jd_keywords(jd_text, top_n=15):
    stop = {
        "and","the","to","of","in","a","is","for","with","we","are","you",
        "on","at","be","will","have","this","that","an","as","by","or",
        "from","it","our","your","was","were","has","had","been","they",
        "can","which","who","also","using","used","like","both","good",
        "well","able","strong","work","must","should","their","ensure",
        "ability","skills","experience","knowledge","candidate","team",
        "looking","join","role","position","apply","applications"
    }
    words = re.findall(r'\b[a-zA-Z]{4,}\b', jd_text.lower())
    freq = {}
    for w in words:
        if w not in stop:
            freq[w] = freq.get(w, 0) + 1
    return [w for w, _ in sorted(freq.items(), key=lambda x: x[1], reverse=True)][:top_n]


# ── Parse resume into sections ────────────────────────────────────────────────
def parse_resume(text):
    """
    Splits resume into sections based on known header keywords.
    Returns list of dicts: [{"label": "OBJECTIVE", "header": "OBJECTIVE:", "body": "..."}]
    """
    # Headers we look for (case-insensitive)
    known = [
        "NAME", "EMAIL", "PHONE", "OBJECTIVE", "SUMMARY",
        "EDUCATION", "SKILLS", "PROJECTS", "INTERNSHIP",
        "EXPERIENCE", "ACHIEVEMENTS", "CERTIFICATIONS",
        "STRENGTHS", "HOBBIES", "LANGUAGES", "DECLARATIONS"
    ]

    pattern = r'^(' + '|'.join(known) + r')\s*[:\-]?\s*(.*)$'
    lines   = text.strip().split('\n')

    sections = []
    current  = None

    for line in lines:
        stripped = line.strip()
        if not stripped:
            if current:
                current["body_lines"].append("")
            continue

        match = re.match(pattern, stripped, re.IGNORECASE)
        if match:
            if current:
                sections.append(current)
            label      = match.group(1).upper()
            rest       = match.group(2).strip()
            current    = {
                "label":      label,
                "header":     stripped.split(rest)[0].strip() if rest else stripped,
                "body_lines": [rest] if rest else []
            }
        else:
            if current:
                current["body_lines"].append(stripped)
            else:
                # Text before any header — treat as preamble
                sections.append({
                    "label":      "PREAMBLE",
                    "header":     "",
                    "body_lines": [stripped]
                })

    if current:
        sections.append(current)

    return sections


# ── Sections to keep exactly as-is (no rewriting) ────────────────────────────
KEEP_AS_IS = {"NAME", "EMAIL", "PHONE", "EDUCATION", "HOBBIES", "PREAMBLE"}


# ── Main pipeline ─────────────────────────────────────────────────────────────
def generate_optimized_resume(resume_text, job_role, job_description):

    print("📌 Extracting keywords from JD...")
    if job_description.strip():
        jd_keywords = get_jd_keywords(job_description)
    else:
        # Fallback keywords based on role
        jd_keywords = get_jd_keywords(job_role + " " + job_role * 5)
    print(f"   Keywords: {', '.join(jd_keywords[:10])}")

    print("\n📂 Parsing resume sections...")
    sections = parse_resume(resume_text)
    for s in sections:
        body = " ".join(s["body_lines"]).strip()
        print(f"   {'✅' if body else '⬜'} {s['label']} — {len(body)} chars")

    print("\n✍️  Rewriting sections...")
    output_parts = []

    for sec in sections:
        label      = sec["label"]
        header     = sec["header"]
        body_lines = sec["body_lines"]
        body       = "\n".join(body_lines).strip()

        if label in KEEP_AS_IS or not body.strip():
            # Print as-is
            if header:
                output_parts.append(header)
            if body:
                output_parts.append(body)
            output_parts.append("")
            print(f"   ⬜ {label} — kept as-is")

        else:
            # Rewrite this section
            print(f"   ✍️  {label} — rewriting...", end=" ", flush=True)
            rewritten = rewrite_section(label, body, job_role, jd_keywords)
            if header:
                output_parts.append(header)
            output_parts.append(rewritten)
            output_parts.append("")
            print("✅")

    final = "\n".join(output_parts).strip()

    print("\n" + "=" * 62)
    print("         ✨ YOUR OPTIMIZED RESUME ✨")
    print("=" * 62)
    print()
    print(final)
    print()
    print("=" * 62)
    print("  ✅ Done! Copy the resume above. Good luck! 🚀")
    print("=" * 62)


print("✅ All functions ready!")
```

```python
display(HTML("""
<div style="background:#1a1a2e;padding:20px 26px;border-radius:14px;margin-bottom:14px">
  <h2 style="color:#e0aaff;font-family:monospace;margin:0;font-size:20px">🤖 AI Resume Rewriter</h2>
  <p style="color:#a8dadc;margin:6px 0 0;font-family:monospace;font-size:13px">
    Paste or upload your resume → enter job details → get optimized resume
  </p>
</div>
"""))

# ── Resume Input ──────────────────────────────────────────────────────────────
display(HTML("<b style='font-family:monospace;font-size:14px'>📄 Step 1: Your Resume</b>"))

input_toggle = widgets.ToggleButtons(
    options=["📝 Paste Text", "📄 Upload PDF"],
    description="Input:",
    style={"description_width": "60px", "button_width": "140px"}
)

resume_box = widgets.Textarea(
    placeholder="Paste your resume text here...",
    layout=widgets.Layout(width="700px", height="240px")
)

pdf_store    = {"text": ""}
upload_btn   = widgets.Button(
    description="📂 Choose PDF",
    button_style="info",
    layout=widgets.Layout(width="180px", display="none")
)
upload_label = widgets.Label("", layout=widgets.Layout(display="none"))
upload_out   = widgets.Output()

def on_upload(b):
    with upload_out:
        clear_output()
        uploaded = files.upload()
        for fname, content in uploaded.items():
            reader = PyPDF2.PdfReader(io.BytesIO(content))
            text = "\n".join(p.extract_text() or "" for p in reader.pages)
            pdf_store["text"] = text
            upload_label.value = f"✅ '{fname}' loaded — {len(text)} chars"

upload_btn.on_click(on_upload)

def on_toggle(change):
    if change["new"] == "📝 Paste Text":
        resume_box.layout.display   = "flex"
        upload_btn.layout.display   = "none"
        upload_label.layout.display = "none"
    else:
        resume_box.layout.display   = "none"
        upload_btn.layout.display   = "flex"
        upload_label.layout.display = "flex"

input_toggle.observe(on_toggle, names="value")

# ── Job Role ──────────────────────────────────────────────────────────────────
role_box = widgets.Text(
    placeholder="e.g. Software Tester, AI Engineer, Data Analyst",
    description="Job Role:",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="700px")
)

# ── Job Description ───────────────────────────────────────────────────────────
jd_box = widgets.Textarea(
    placeholder="Paste the job description here (optional — but gives better results)...",
    layout=widgets.Layout(width="700px", height="180px")
)

# ── Generate Button ───────────────────────────────────────────────────────────
gen_btn = widgets.Button(
    description="✨ Generate Optimized Resume",
    button_style="success",
    layout=widgets.Layout(width="280px", height="44px", margin="14px 0 0 0")
)
result_out = widgets.Output()

def on_generate(b):
    with result_out:
        clear_output()

        # Get resume text
        if input_toggle.value == "📝 Paste Text":
            resume_text = resume_box.value.strip()
            if not resume_text:
                print("❌ Please paste your resume text.")
                return
        else:
            resume_text = pdf_store["text"].strip()
            if not resume_text:
                print("❌ Please upload a PDF file first.")
                return

        job_role = role_box.value.strip()
        if not job_role:
            print("❌ Please enter a target job role.")
            return

        job_description = jd_box.value.strip()

        print("⏳ Starting... please wait (30–60 sec)\n")

        try:
            generate_optimized_resume(resume_text, job_role, job_description)
        except Exception as e:
            print(f"❌ Error: {e}")

gen_btn.on_click(on_generate)

# ── Layout ────────────────────────────────────────────────────────────────────
display(widgets.VBox([
    input_toggle,
    resume_box,
    upload_btn,
    upload_label,
    upload_out,
    widgets.HTML("<br><b style='font-family:monospace;font-size:14px'>🎯 Step 2: Target Job Role</b>"),
    role_box,
    widgets.HTML("<br><b style='font-family:monospace;font-size:14px'>📋 Step 3: Job Description <span style='color:gray;font-weight:normal'>(optional)</span></b>"),
    jd_box,
    gen_btn,
    result_out
], layout=widgets.Layout(padding="10px", gap="4px")))
```
