```python
# ╔══════════════════════════════════════════════════════════════╗
# ║           AI RESUME REWRITER — Mistral-7B                   ║
# ║  ⚠️  Set Runtime → T4 GPU before running this cell          ║
# ╚══════════════════════════════════════════════════════════════╝

# ── STEP 1: Install ──────────────────────────────────────────────
import subprocess
subprocess.run(["pip", "install", "transformers", "torch", "accelerate",
                "bitsandbytes", "PyPDF2", "scikit-learn", "sentencepiece",
                "huggingface_hub", "-q"], check=True)

# ── STEP 2: Imports ──────────────────────────────────────────────
import re, io, torch
import PyPDF2
import ipywidgets as widgets
import numpy as np
from IPython.display import display, clear_output, HTML
from google.colab import files
from huggingface_hub import login
from sklearn.feature_extraction.text import TfidfVectorizer
from transformers import (AutoTokenizer, AutoModelForCausalLM,
                          BitsAndBytesConfig)

# ── STEP 3: HuggingFace Login ────────────────────────────────────
# Regenerate your token at huggingface.co/settings/tokens
# and paste it below
HF_TOKEN = "Hugging_Face_Token"
login(HF_TOKEN, add_to_git_credential=False)
print("✅ Logged in to HuggingFace!")

# ── STEP 4: Load Mistral-7B (4-bit, fits free Colab T4 GPU) ─────
print("⏳ Loading Mistral-7B-Instruct... (3–5 min first time)")

MODEL_ID   = "mistralai/Mistral-7B-Instruct-v0.2"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, token=HF_TOKEN)
model     = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    quantization_config=bnb_config,
    device_map="auto",
    token=HF_TOKEN
)
print("✅ Mistral-7B ready!\n")

# ── STEP 5: Helper Functions ─────────────────────────────────────

def extract_tfidf_keywords(jd_text, top_n=15):
    """Pull best keywords from JD using TF-IDF + bigrams."""
    stop = {
        "and","the","to","of","in","a","is","for","with","we","are","you",
        "on","at","be","will","have","this","that","an","as","by","or",
        "from","it","our","your","was","were","has","had","been","they",
        "can","which","who","also","using","used","like","both","good",
        "well","able","strong","work","must","should","their","ensure",
        "ability","candidate","team","looking","join","role","position",
        "apply","required","preferred","including","plus","candidate"
    }
    try:
        vec    = TfidfVectorizer(stop_words=list(stop),
                                 ngram_range=(1, 2), min_df=1)
        mat    = vec.fit_transform([jd_text])
        scores = np.array(mat.todense()).flatten()
        vocab  = vec.get_feature_names_out()
        ranked = sorted(zip(vocab, scores), key=lambda x: x[1], reverse=True)
        return [w for w, s in ranked if s > 0][:top_n]
    except Exception:
        words = re.findall(r'\b[a-zA-Z]{4,}\b', jd_text.lower())
        freq  = {}
        for w in words:
            if w not in stop:
                freq[w] = freq.get(w, 0) + 1
        return sorted(freq, key=freq.get, reverse=True)[:top_n]


def mistral_generate(prompt, max_new_tokens=1400):
    """Run Mistral inference with instruct format."""
    formatted = f"[INST] {prompt.strip()} [/INST]"
    inputs    = tokenizer(formatted, return_tensors="pt").to(model.device)
    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            repetition_penalty=1.15,
            pad_token_id=tokenizer.eos_token_id
        )
    generated = output[0][inputs["input_ids"].shape[1]:]
    return tokenizer.decode(generated, skip_special_tokens=True).strip()


def generate_optimized_resume(resume_text, job_role, job_description):
    """Full pipeline: extract keywords → build prompt → run Mistral → print output."""

    # TF-IDF keywords
    jd_src   = job_description if job_description.strip() else job_role
    keywords = extract_tfidf_keywords(jd_src)
    kw_str   = ", ".join(keywords[:12])
    print(f"📌 JD Keywords (TF-IDF): {kw_str}\n")

    # JD block for prompt
    jd_block = (
        f"JOB DESCRIPTION:\n{job_description.strip()}"
        if job_description.strip()
        else f"Target role: {job_role}"
    )

    # The prompt
    prompt = f"""You are an expert professional resume writer with 10 years of experience helping candidates land jobs.

Your task: Rewrite the resume below so it is polished, ATS-optimized, and perfectly tailored to the job role.

RULES — follow every rule strictly:
1. Preserve EVERY section heading exactly as in the original: NAME, EMAIL, PHONE, OBJECTIVE, EDUCATION, SKILLS, PROJECTS, INTERNSHIP, ACHIEVEMENTS, STRENGTHS, HOBBIES.
2. Keep all real personal facts unchanged: full name, email, phone number, college name, CGPA, company names, project names.
3. Rewrite OBJECTIVE to be sharp, role-specific, and mention the job title "{job_role}".
4. Rewrite SKILLS: keep all original skills and naturally add relevant ones from these JD keywords — {kw_str}.
5. Rewrite each PROJECT: keep the project title exactly, rewrite the description using strong action verbs and measurable outcomes. Add testing/relevant terminology from the JD.
6. Rewrite INTERNSHIP: keep company name exactly, rewrite duties with strong action verbs, add JD-relevant impact.
7. Rewrite ACHIEVEMENTS: make each point crisp and impactful with a bullet point.
8. Rewrite STRENGTHS: make them relevant to the job role "{job_role}".
9. Keep EDUCATION and HOBBIES word-for-word exactly as in the original. Do not change them at all.
10. Use strong action verbs throughout: developed, designed, implemented, executed, validated, tested, built, optimized, delivered, collaborated.
11. Do NOT invent fake experience, fake certifications, or fake companies.
12. Output ONLY the rewritten resume — no explanations, no commentary, no notes before or after.

{jd_block}

ORIGINAL RESUME:
{resume_text}

REWRITTEN RESUME:"""

    print("✍️  Mistral-7B is rewriting your resume... (1–2 min)\n")
    result = mistral_generate(prompt)

    # Strip any leftover prompt artifacts
    for tag in ["REWRITTEN RESUME:", "[INST]", "[/INST]", "OUTPUT:", "Note:", "Note :"]:
        result = result.replace(tag, "").strip()

    # Print final output
    print("\n" + "=" * 64)
    print("            ✨  YOUR OPTIMIZED RESUME  ✨")
    print("=" * 64 + "\n")
    print(result)
    print("\n" + "=" * 64)
    print("  ✅ Done! Copy everything between the lines above.")
    print("=" * 64)


# ── STEP 6: UI ───────────────────────────────────────────────────

display(HTML("""
<div style="background:#1a1a2e;padding:20px 26px;border-radius:14px;margin-bottom:16px">
  <h2 style="color:#e0aaff;font-family:monospace;margin:0;font-size:20px">
    🤖 AI Resume Rewriter — Mistral-7B
  </h2>
  <p style="color:#a8dadc;margin:8px 0 0;font-family:monospace;font-size:13px">
    Paste or upload resume → enter job role + JD → click Generate
  </p>
</div>
"""))

# Resume input toggle
display(HTML("<b style='font-family:monospace;font-size:14px'>📄 Step 1: Your Resume</b>"))

input_toggle = widgets.ToggleButtons(
    options=["📝 Paste Text", "📄 Upload PDF"],
    description="Input:",
    style={"description_width": "60px", "button_width": "145px"}
)

resume_box = widgets.Textarea(
    placeholder="Paste your full resume text here...",
    layout=widgets.Layout(width="720px", height="250px")
)

pdf_store    = {"text": ""}
upload_btn   = widgets.Button(
    description="📂 Choose PDF File",
    button_style="info",
    layout=widgets.Layout(width="190px", display="none")
)
upload_label = widgets.Label("", layout=widgets.Layout(display="none"))
upload_out   = widgets.Output()

def on_upload(b):
    with upload_out:
        clear_output()
        uploaded = files.upload()
        for fname, content in uploaded.items():
            reader = PyPDF2.PdfReader(io.BytesIO(content))
            text   = "\n".join(p.extract_text() or "" for p in reader.pages)
            pdf_store["text"]  = text
            upload_label.value = f"✅ '{fname}' loaded — {len(text)} characters extracted"

upload_btn.on_click(on_upload)

def on_toggle(change):
    if change["new"] == "📝 Paste Text":
        resume_box.layout.display   = "flex"
        upload_btn.layout.display   = "none"
        upload_label.layout.display = "none"
    else:
        resume_box.layout.display   = "none"
        upload_btn.layout.display   = "flex"
        upload_label.layout.display = "flex"

input_toggle.observe(on_toggle, names="value")

# Job role
role_box = widgets.Text(
    placeholder="e.g. Software Tester, AI Engineer, Data Analyst, Web Developer",
    description="Job Role:",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="720px")
)

# Job description
jd_box = widgets.Textarea(
    placeholder="Paste the job description here (optional but strongly recommended for best results)...",
    layout=widgets.Layout(width="720px", height="190px")
)

# Generate button + output
gen_btn    = widgets.Button(
    description="✨ Generate Optimized Resume",
    button_style="success",
    layout=widgets.Layout(width="290px", height="46px", margin="16px 0 0 0")
)
result_out = widgets.Output()

def on_generate(b):
    with result_out:
        clear_output()

        # Validate resume input
        if input_toggle.value == "📝 Paste Text":
            resume_text = resume_box.value.strip()
            if not resume_text:
                print("❌ Please paste your resume text first.")
                return
        else:
            resume_text = pdf_store["text"].strip()
            if not resume_text:
                print("❌ Please upload a PDF file first.")
                return

        # Validate job role
        job_role = role_box.value.strip()
        if not job_role:
            print("❌ Please enter your target job role.")
            return

        job_description = jd_box.value.strip()

        print("⏳ Starting pipeline... please wait (1–2 minutes)\n")

        try:
            generate_optimized_resume(resume_text, job_role, job_description)
        except torch.cuda.OutOfMemoryError:
            print("❌ GPU out of memory.")
            print("👉 Go to Runtime → Restart runtime, then run the cell again.")
        except Exception as e:
            print(f"❌ Error: {e}")
            print("👉 Make sure Runtime is set to T4 GPU.")

gen_btn.on_click(on_generate)

# Render UI
display(widgets.VBox([
    input_toggle,
    resume_box,
    upload_btn,
    upload_label,
    upload_out,
    widgets.HTML("<br><b style='font-family:monospace;font-size:14px'>🎯 Step 2: Target Job Role</b>"),
    role_box,
    widgets.HTML("<br><b style='font-family:monospace;font-size:14px'>📋 Step 3: Job Description <span style='color:gray;font-weight:normal'>(optional — but gives much better output)</span></b>"),
    jd_box,
    gen_btn,
    result_out
], layout=widgets.Layout(padding="10px", gap="6px")))
```
