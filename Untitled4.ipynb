```python
import re, io, torch
import PyPDF2
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
from google.colab import files
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
```

```python
MODEL_ID   = "mistralai/Mistral-7B-Instruct-v0.2"

bnb_config = BitsAndBytesConfig(
    load_in_4bit              = True,
    bnb_4bit_use_double_quant = True,
    bnb_4bit_quant_type       = "nf4",
    bnb_4bit_compute_dtype    = torch.float16
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model     = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    quantization_config = bnb_config,
    device_map          = "auto"
)
```

```python
def optimize_resume(resume_text):

    prompt = f"""You are a professional resume writer with 10+ years of experience.

Your task is to optimize the resume below to make it highly professional, impactful, and ATS-friendly.

STRICT RULES — follow every single one:
1. Keep the EXACT same section headings as the original (NAME, EMAIL, PHONE, OBJECTIVE, EDUCATION, SKILLS, PROJECTS, INTERNSHIP, ACHIEVEMENTS, STRENGTHS, HOBBIES etc).
2. Keep all personal facts unchanged — full name, email, phone, college name, CGPA, company names, project names, certification names.
3. OBJECTIVE: Rewrite it to be sharp, confident, and professional. Make it sound like a strong candidate.
4. SKILLS: Keep all original skills. Add relevant professional skills that naturally fit the candidate's background.
5. PROJECTS: Keep each project title exactly. Rewrite descriptions using strong action verbs and professional language. Show impact wherever possible.
6. INTERNSHIP: Keep company name exactly. Rewrite duties professionally with strong action verbs.
7. ACHIEVEMENTS: Make each point crisp, confident, and impactful using bullet points.
8. STRENGTHS: Rewrite to sound professional and relevant to a technical/software role.
9. EDUCATION and HOBBIES: Keep exactly word-for-word. Do not change at all.
10. Use strong action verbs: developed, designed, implemented, built, optimized, delivered, collaborated, executed, validated, engineered.
11. Do NOT invent fake experience, fake companies, fake certifications, or fake skills.
12. Do NOT add any explanation, notes, or commentary. Output ONLY the optimized resume.
13. Keep the same formatting style — if original uses numbered lists for projects, keep numbered lists.

ORIGINAL RESUME:
{resume_text}

OPTIMIZED RESUME:"""

    # Tokenize
    formatted = f"[INST] {prompt.strip()} [/INST]"
    inputs    = tokenizer(
        formatted,
        return_tensors = "pt",
        truncation     = True,
        max_length     = 3000
    ).to(model.device)

    # Generate
    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens    = 1400,
            do_sample         = False,
            repetition_penalty = 1.15,
            pad_token_id      = tokenizer.eos_token_id
        )

    # Decode only the new tokens (not the prompt)
    generated = output[0][inputs["input_ids"].shape[1]:]
    result    = tokenizer.decode(generated, skip_special_tokens=True).strip()

    # Clean any leftover prompt artifacts
    for tag in ["OPTIMIZED RESUME:", "REWRITTEN RESUME:", "[INST]", "[/INST]",
                "OUTPUT:", "Note:", "Note :", "Here is", "Here's"]:
        result = result.replace(tag, "").strip()

    return result
```

```python
!pip install reportlab PyPDF2 -q

from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import mm
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, HRFlowable
import re, io
import PyPDF2
from google.colab import files

# ── Step 1: Choose input method ───────────────────────────────────
print("=" * 55)
print("        HOW DO YOU WANT TO PROVIDE YOUR RESUME?")
print("=" * 55)
print("  1 → Paste text")
print("  2 → Upload PDF")
print("-" * 55)

choice      = input("Enter 1 or 2: ").strip()
resume_text = ""

if choice == "1":
    print("\nPaste your resume below.")
    print("When done, type END on a new line and press Enter:\n")
    lines = []
    while True:
        line = input()
        if line.strip().upper() == "END":
            break
        lines.append(line)
    resume_text = "\n".join(lines).strip()
    print(f"\n✅ Resume received — {len(resume_text)} characters")

elif choice == "2":
    print("\n📂 Select your PDF file...")
    uploaded = files.upload()
    for fname, content in uploaded.items():
        reader      = PyPDF2.PdfReader(io.BytesIO(content))
        resume_text = "\n".join(p.extract_text() or "" for p in reader.pages).strip()
        print(f"✅ '{fname}' loaded — {len(resume_text)} characters")

else:
    print("❌ Invalid choice. Re-run this cell and enter 1 or 2.")

# ── Step 2: Optimize ──────────────────────────────────────────────
if resume_text:

    print("\n⏳ Optimizing your resume... (1–2 minutes)\n")

    prompt = f"""You are a professional resume writer with 10+ years of experience.

Optimize the resume below to make it highly professional, impactful, and ATS-friendly.

STRICT RULES:
1. Keep the EXACT same section headings as the original (NAME, EMAIL, PHONE, OBJECTIVE, EDUCATION, SKILLS, PROJECTS, INTERNSHIP, ACHIEVEMENTS, STRENGTHS, HOBBIES etc).
2. Keep all personal facts unchanged — full name, email, phone, college name, CGPA, company names, project names, certification names.
3. OBJECTIVE: Rewrite it to be sharp, confident, and professional. Make it sound like a strong candidate.
4. SKILLS: Keep all original skills. Add relevant professional skills that naturally fit the candidate's background.
5. PROJECTS: Keep each project title exactly. Rewrite descriptions using strong action verbs and professional language. Show impact wherever possible.
6. INTERNSHIP: Keep company name exactly. Rewrite duties professionally with strong action verbs.
7. ACHIEVEMENTS: Make each point crisp, confident, and impactful using bullet points.
8. STRENGTHS: Rewrite to sound professional and relevant to a technical/software role.
9. EDUCATION and HOBBIES: Keep exactly word-for-word. Do not change at all.
10. Use strong action verbs: developed, designed, implemented, built, optimized, delivered, collaborated, executed, validated, engineered.
11. Do NOT invent fake experience, fake companies, fake certifications, or fake skills.
12. Do NOT add any explanation, notes, or commentary. Output ONLY the optimized resume.
13. Keep the same formatting style — if original uses numbered lists for projects, keep numbered lists.

ORIGINAL RESUME:
{resume_text}

OPTIMIZED RESUME:"""

    formatted = f"[INST] {prompt.strip()} [/INST]"
    inputs    = tokenizer(
        formatted,
        return_tensors = "pt",
        truncation     = True,
        max_length     = 3000
    ).to(model.device)

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens     = 1400,
            do_sample          = False,
            repetition_penalty = 1.15,
            pad_token_id       = tokenizer.eos_token_id
        )

    generated = output[0][inputs["input_ids"].shape[1]:]
    result    = tokenizer.decode(generated, skip_special_tokens=True).strip()

    for tag in ["OPTIMIZED RESUME:", "REWRITTEN RESUME:", "[INST]", "[/INST]",
                "OUTPUT:", "Note:", "Here is", "Here's"]:
        result = result.replace(tag, "").strip()

    # ── Step 3: Print output ──────────────────────────────────────
    print("=" * 62)
    print("           ✨  YOUR OPTIMIZED RESUME  ✨")
    print("=" * 62)
    print()
    print(result)
    print()
    print("=" * 62)
    print("  ✅ Optimization complete!")
    print("=" * 62)

    # ── Step 4: Download as PDF ───────────────────────────────────
    print()
    download_choice = input("📥 Do you want to download as PDF? (yes / no): ").strip().lower()

    if download_choice in ["yes", "y"]:

        pdf_filename = "optimized_resume.pdf"

        doc = SimpleDocTemplate(
            pdf_filename,
            pagesize     = A4,
            leftMargin   = 20 * mm,
            rightMargin  = 20 * mm,
            topMargin    = 18 * mm,
            bottomMargin = 18 * mm
        )

        # ── Styles ────────────────────────────────────────────────
        style_heading = ParagraphStyle(
            "Heading",
            fontName    = "Helvetica-Bold",
            fontSize    = 12,
            textColor   = colors.HexColor("#1a1a2e"),
            spaceAfter  = 2,
            spaceBefore = 8,
        )
        style_body = ParagraphStyle(
            "Body",
            fontName  = "Helvetica",
            fontSize  = 10,
            textColor = colors.HexColor("#333333"),
            spaceAfter= 3,
            leading   = 14,
            wordWrap  = "CJK"
        )
        style_bullet = ParagraphStyle(
            "Bullet",
            fontName     = "Helvetica",
            fontSize     = 10,
            textColor    = colors.HexColor("#333333"),
            spaceAfter   = 3,
            leading      = 14,
            leftIndent   = 12,
            bulletIndent = 0,
            wordWrap     = "CJK"
        )
        style_numbered = ParagraphStyle(
            "Numbered",
            fontName    = "Helvetica-Bold",
            fontSize    = 10,
            textColor   = colors.HexColor("#1a1a2e"),
            spaceAfter  = 2,
            spaceBefore = 6,
            wordWrap    = "CJK"
        )

        # ── Section heading keywords ──────────────────────────────
        section_keywords = [
            "NAME:", "EMAIL:", "PHONE:", "OBJECTIVE:", "EDUCATION:",
            "SKILLS:", "PROJECTS:", "INTERNSHIP:", "EXPERIENCE:",
            "ACHIEVEMENTS:", "STRENGTHS:", "HOBBIES:", "CERTIFICATIONS:",
            "LANGUAGES:", "DECLARATION:", "SUMMARY:", "TECHNICAL SKILLS:",
            "PROFESSIONAL SUMMARY"
        ]

        # ── Build PDF story ───────────────────────────────────────
        story = []

        for line in result.split("\n"):
            stripped = line.strip()

            if not stripped:
                story.append(Spacer(1, 4))
                continue

            # Escape XML special chars
            safe = (stripped
                    .replace("&", "&amp;")
                    .replace("<", "&lt;")
                    .replace(">", "&gt;"))

            is_heading = any(
                stripped.upper().startswith(kw) for kw in section_keywords
            )

            if is_heading:
                story.append(Spacer(1, 4))
                story.append(Paragraph(safe, style_heading))
                story.append(HRFlowable(
                    width     = "100%",
                    thickness = 0.5,
                    color     = colors.HexColor("#999999"),
                    spaceAfter= 4
                ))

            elif stripped.startswith(("•", "-", "*")):
                story.append(Paragraph(f"&nbsp;&nbsp;&nbsp;{safe}", style_bullet))

            elif re.match(r'^\d+\.', stripped):
                story.append(Paragraph(safe, style_numbered))

            else:
                story.append(Paragraph(safe, style_body))

        # ── Save & Download ───────────────────────────────────────
        doc.build(story)
        print(f"\n📄 Downloading '{pdf_filename}'...")
        files.download(pdf_filename)
        print("✅ PDF downloaded successfully!")

    else:
        print("\n👍 Okay! Copy the resume printed above.")
```
