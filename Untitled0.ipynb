```python
import re, textwrap, io
from transformers import T5ForConditionalGeneration, T5Tokenizer
from sentence_transformers import SentenceTransformer, util
import PyPDF2
from google.colab import files
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

print("✅ Imports done!")
```

```python
print("⏳ Loading FLAN-T5...")
flan_tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")
flan_model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-base")
print("✅ FLAN-T5 ready!")

print("⏳ Loading MiniLM...")
minilm_model = SentenceTransformer("all-MiniLM-L6-v2")
print("✅ MiniLM ready!")

print("\n🎉 All models loaded!")
```

```python
# ── Similarity Score ─────────────────────────────────────────────────────────
def calculate_similarity(text1, text2):
    e1 = minilm_model.encode(text1, convert_to_tensor=True)
    e2 = minilm_model.encode(text2, convert_to_tensor=True)
    return round(util.cos_sim(e1, e2).item() * 100, 2)

# ── Keyword Extractor ─────────────────────────────────────────────────────────
def extract_keywords(text, top_n=20):
    stop = {
        "and","the","to","of","in","a","is","for","with","we","are","you","on",
        "at","be","will","have","this","that","an","as","by","or","from","it",
        "our","your","was","were","has","had","been","they","can","which","who",
        "also","using","used","like","both","good","well","able","strong","work"
    }
    words = re.findall(r'\b[a-zA-Z]{4,}\b', text.lower())
    freq = {}
    for w in words:
        if w not in stop:
            freq[w] = freq.get(w, 0) + 1
    return sorted(freq, key=freq.get, reverse=True)[:top_n]

# ── Parse Resume into Named Sections ─────────────────────────────────────────
def parse_sections(text):
    """
    Extracts each section as a block, preserving all content under it.
    Returns ordered list of (heading_line, body_text) tuples.
    """
    # Known section header patterns
    header_patterns = [
        r'^(NAME\s*:.+)$',
        r'^(EMAIL\s*:.+)$',
        r'^(PHONE\s*:.+)$',
        r'^(OBJECTIVE\s*:.*)$',
        r'^(EDUCATION\s*:.*)$',
        r'^(SKILLS\s*:.*)$',
        r'^(PROJECTS\s*:.*)$',
        r'^(INTERNSHIP\s*:.*)$',
        r'^(ACHIEVEMENTS\s*:.*)$',
        r'^(STRENGTHS\s*:.*)$',
        r'^(HOBBIES\s*:.*)$',
        r'^(EXPERIENCE\s*:.*)$',
        r'^(CERTIFICATIONS\s*:.*)$',
    ]

    lines = text.strip().split('\n')
    sections = []  # list of {'header': str, 'body': [lines]}
    current = None

    for line in lines:
        stripped = line.strip()
        if not stripped:
            if current:
                current['body'].append('')
            continue

        matched = False
        for pat in header_patterns:
            if re.match(pat, stripped, re.IGNORECASE):
                if current:
                    sections.append(current)
                current = {'header': stripped, 'body': []}
                matched = True
                break

        if not matched:
            if current:
                current['body'].append(stripped)
            else:
                # Lines before any known header
                sections.append({'header': stripped, 'body': []})

    if current:
        sections.append(current)

    return sections

# ── Rewrite a Single Section with FLAN-T5 ─────────────────────────────────────
def rewrite_section_flan(section_name, section_content, job_role, jd_keywords):
    """
    Rewrites one section's content using FLAN-T5.
    Keeps the prompt tight so model doesn't wander.
    """
    if not section_content.strip():
        return section_content

    kw_hint = ", ".join(jd_keywords[:10])
    text = section_content[:350]  # stay within token limit

    prompt = (
        f"Rewrite this resume {section_name} section for a {job_role} job. "
        f"Preserve the exact structure and all original facts. "
        f"Use action verbs. Add relevant keywords: {kw_hint}. "
        f"Do not add fake experience. Keep it concise.\n"
        f"Original:\n{text}\n"
        f"Rewritten:"
    )

    inputs = flan_tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True)
    outputs = flan_model.generate(
        inputs["input_ids"],
        max_new_tokens=220,
        num_beams=4,
        early_stopping=True,
        no_repeat_ngram_size=3
    )
    result = flan_tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    return result

# ── Sections that should NOT be rewritten (keep as-is) ───────────────────────
SKIP_REWRITE = {"name", "email", "phone", "education", "hobbies", "achievements", "certifications"}

# ── Main Pipeline ─────────────────────────────────────────────────────────────
def run_pipeline(resume_text, job_role, job_description):
    print("\n" + "="*60)
    print("  🔄 PROCESSING YOUR RESUME...")
    print("="*60)

    # Default JD if empty
    if not job_description.strip():
        job_description = (
            f"We are looking for a skilled {job_role} with strong technical "
            f"and communication skills."
        )

    # Extract JD keywords
    jd_keywords = extract_keywords(job_description)
    print(f"\n📌 JD Keywords detected: {', '.join(jd_keywords[:10])}")

    # Parse resume into sections
    sections = parse_sections(resume_text)
    print(f"📂 Sections found: {len(sections)}")

    # Rewrite sections
    print("\n✍️  Rewriting sections...")
    output_lines = []

    for sec in sections:
        header      = sec['header']
        body_lines  = sec['body']
        body_text   = '\n'.join(body_lines).strip()

        sec_key = header.split(':')[0].strip().lower()

        if sec_key in SKIP_REWRITE or not body_text:
            output_lines.append(header)
            if body_lines:
                output_lines.extend(body_lines)
            output_lines.append('')
            print(f"   ⬜ {sec_key.upper()} — kept as-is")
        else:
            print(f"   ✍️  {sec_key.upper()} — rewriting...", end=" ")
            rewritten_body = rewrite_section_flan(sec_key, body_text, job_role, jd_keywords)
            output_lines.append(header)
            output_lines.append(rewritten_body)
            output_lines.append('')
            print("✅")

    # Final resume
    final_resume = '\n'.join(output_lines).strip()

    print("\n" + "="*60)
    print("       ✨ YOUR REWRITTEN RESUME ✨")
    print("="*60 + "\n")
    print(final_resume)

    print("\n" + "="*60)
    print("  ✅ Done! Good luck 🚀")
    print("="*60)

print("✅ All functions ready!")
```

```python
display(HTML("""
<div style="background:#1e1e2e;padding:18px 24px;border-radius:12px;margin-bottom:10px">
  <h2 style="color:#cba6f7;font-family:monospace;margin:0">🤖 AI Resume Rewriter</h2>
  <p style="color:#a6e3a1;margin:6px 0 0;font-family:monospace;font-size:13px">
    Fill the fields below and click <b>Generate Rewritten Resume</b>
  </p>
</div>
"""))

# ── Toggle: Text or PDF ───────────────────────────────────────────────────────
input_type = widgets.ToggleButtons(
    options=["📝 Paste Text", "📄 Upload PDF"],
    description="Resume:",
    style={"description_width": "70px", "button_width": "140px"}
)

# ── Text Area ─────────────────────────────────────────────────────────────────
resume_textarea = widgets.Textarea(
    placeholder="Paste your resume text here...",
    layout=widgets.Layout(width="700px", height="230px")
)

# ── PDF Upload ────────────────────────────────────────────────────────────────
upload_btn    = widgets.Button(description="📂 Choose PDF File", button_style="info",
                               layout=widgets.Layout(width="200px", display="none"))
upload_status = widgets.Label(value="", layout=widgets.Layout(display="none"))
upload_out    = widgets.Output()
pdf_store     = {"text": ""}

def on_upload(b):
    with upload_out:
        clear_output()
        print("📂 Opening file picker...")
        uploaded = files.upload()
        for fname, content in uploaded.items():
            reader = PyPDF2.PdfReader(io.BytesIO(content))
            text = "\n".join(p.extract_text() or "" for p in reader.pages)
            pdf_store["text"] = text
            upload_status.value = f"✅ '{fname}' loaded — {len(text)} characters"
            print(upload_status.value)

upload_btn.on_click(on_upload)

def toggle_input(change):
    if change["new"] == "📝 Paste Text":
        resume_textarea.layout.display = "flex"
        upload_btn.layout.display      = "none"
        upload_status.layout.display   = "none"
    else:
        resume_textarea.layout.display = "none"
        upload_btn.layout.display      = "flex"
        upload_status.layout.display   = "flex"

input_type.observe(toggle_input, names="value")

# ── Job Role ──────────────────────────────────────────────────────────────────
job_role_box = widgets.Text(
    placeholder="e.g. Software Tester, AI Engineer, Data Analyst",
    description="Job Role:",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="700px")
)

# ── Job Description ───────────────────────────────────────────────────────────
jd_box = widgets.Textarea(
    placeholder="Paste the job description here (optional but improves scoring)...",
    layout=widgets.Layout(width="700px", height="160px")
)

# ── Run Button ────────────────────────────────────────────────────────────────
run_btn = widgets.Button(
    description="🚀 Generate Rewritten Resume",
    button_style="success",
    layout=widgets.Layout(width="310px", height="44px", margin="12px 0 0 0")
)
run_out = widgets.Output()

def on_run(b):
    with run_out:
        clear_output()

        # Get resume text
        if input_type.value == "📝 Paste Text":
            resume_text = resume_textarea.value.strip()
            if not resume_text:
                print("❌ Please paste your resume text.")
                return
        else:
            resume_text = pdf_store["text"].strip()
            if not resume_text:
                print("❌ Please upload a PDF first.")
                return

        job_role = job_role_box.value.strip()
        if not job_role:
            print("❌ Please enter a target job role.")
            return

        job_description = jd_box.value.strip()
        run_pipeline(resume_text, job_role, job_description)

run_btn.on_click(on_run)

# ── Display ───────────────────────────────────────────────────────────────────
display(widgets.VBox([
    widgets.HTML("<b style='font-family:monospace'>📄 Resume Input</b>"),
    input_type,
    resume_textarea,
    upload_btn,
    upload_status,
    upload_out,
    widgets.HTML("<br><b style='font-family:monospace'>🎯 Target Job Role</b>"),
    job_role_box,
    widgets.HTML("<br><b style='font-family:monospace'>📋 Job Description <span style=\"color:gray;font-weight:normal\">(optional)</span></b>"),
    jd_box,
    run_btn,
    run_out
], layout=widgets.Layout(padding="10px", gap="6px")))
```
